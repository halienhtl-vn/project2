```python
import pandas as pd
#Load dataset
df = pd.read_csv(r"E:\02. Hochschule Mainz\10. Jobs\2025\Python\SuperStore Sales.csv", encoding='ISO-8859-1')
#View first few rows
df.head()
```

```python
#Check dataset info
df.info()
```

```python
df.describe()
```

```python
#Check missing values
df.isnull().sum()
```

```python
#Average profit (numpy)
import numpy as np
avg_profit=np.mean(df["Profit"])
print("Average Profit:", round(avg_profit,2))
```

```python
# Profit by region
df.groupby("Region")["Profit"].sum().sort_values(ascending=False)
```

```python
#Total Sales by category
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
print(category_sales)
                             
```

```python
#Simple vizualizations (Matplotlib)
#Bar chart: Sales by category

import matplotlib.pyplot as plt

category_sales.plot(kind="bar", color="skyblue")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
# Simple vizualizations (Seaborn)
# Boxplot: Profit distribution by category
import seaborn as sns
sns.boxplot(x="Category", y="Profit", data=df)
plt.title("Profit Distribution by Category")
plt.show()
```
